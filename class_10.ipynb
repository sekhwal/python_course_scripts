```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#add labels
plt.plot(x,y)
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
```

```python
#Change the color of the line in the plot
plt.plot(x,y,color='g',linestyle=":",linewidth=2)
plt.show()
```

```python
#Adding two lines in the same plot 
y1=x*2
y2=x*3
```

```python
y1
```

```python
y2
```

```python
plt.plot(x,y1,color='g',linestyle=":",linewidth=3)
plt.plot(x,y2,color='r',linestyle="-.",linewidth=3)
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.grid(True)
plt.show()
```

```python
#Adding subplots
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x,y2,color="y",linestyle=":",linewidth=3)
plt.grid(True)
plt.show()
```

```python
#Bar plot
std={'SAM':10,'JOHN':20,'KUMAR':30}
```

```python
names=list(std.keys())
values=list(std.values())
```

```python
plt.bar(names,values)
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.grid(True)
plt.show()
```

```python
##Box plot
a=[1,2,3,6,4,5,8]
b=[5,6,7,3,7,8,9]
c=[3,4,5,6,2,8,4]
```

```python
#Pie-Chart
fruit=['Apple','Orange','Mango','Guava']
costs=[45,56,78,54]
```

```python
plt.pie(costs,labels=fruit)
plt.show()
```

```python
#Adding percentage in the pie-chart
plt.pie(costs,labels=fruit,autopct='%0.1f%%',colors=['yellow','pink','green','orange'])
plt.show()
```

```python
#DoughNut-Chart
plt.pie(costs,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```

```python
#SeaBorn Line plot
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint",y='signal',data=fmri)
plt.show()
```

```python
#Grouping data with hue
```

```python
sns.lineplot(x="timepoint",y='signal',hue='event',style="event",markers=True,data=fmri)
plt.show()
```

```python
#Example data
import pandas as pd
sns.set(style="whitegrid")
iris=pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
sns.barplot(x='class',y='sepal_length',data=iris)
plt.show()
```

```python
sns.barplot(x='sepal_width',y='sepal_length',hue='class',data=iris)
plt.show()
```

```python
#different palette set on the bar plot
#sns.barplot(x='class',y='sepal_length',data=iris,palette='Blues_d')
sns.barplot(x='class',y='sepal_length',data=iris,palette='rocket')
#sns.barplot(x='class',y='sepal_length',data=iris,palette='vlag')
plt.show()
```

```python
#Scatterplot with iris dataset
sns.scatterplot(x='sepal_length',y='sepal_width',hue='class',style='class',data=iris)
plt.show()
```

```python
iris.head()
```

```python
#Seaborn jointplot
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='brown',kind='reg')
plt.show()
```

```python
#Seaborn Boxplot
sns.boxplot(x='sepal_length',y='petal_length',palette="Set1",linewidth=3,hue='class',data=iris)
plt.show()
```

```python
#Seaborn pairplot
df=sns.load_dataset('iris')
```

```python
sns.pairplot(df,hue='species')
plt.show()
```
